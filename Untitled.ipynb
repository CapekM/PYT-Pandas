```python
import pandas as pd
from explanations import file_names, main_columns, ped_columns, decode_key
%matplotlib inline
```

```python
# read csv
data = {}
for file_name in file_names:
    data[file_name] = pd.read_csv(f'data/2018/{file_name}', encoding='iso-8859-2', names=main_columns, sep=';', low_memory=False)
    data[file_name].dropna(axis='columns', how='all', inplace=True)
    data[file_name]['Kraj'] = file_names[file_name]

chodci = pd.read_csv(f'data/2018/CHODCI.csv', encoding='iso-8859-2', names=ped_columns, sep=';', low_memory=False)
chodci
```

```python
# make it one table and save
clean = pd.concat( data, sort = False )
clean.to_csv('clean-2018.csv')
clean
```

### Kolik procent zaznamenaných nehod bylo smrtelných?

```python
deadly = clean[clean['usmrceno_osob'] > 0]['usmrceno_osob']
percentage_deadly = len(deadly)/len(clean.index)
print(f"{percentage_deadly*100}%")
```

### Kolik lidí zemřelo při srážce s odrazníkem, patníkem, sloupkem dopravní značky apod.?

```python
solid_object = clean[(clean['druh_pevne_prekazky'] == 3) & deadly]
solid_object['usmrceno_osob'].sum()
```

### Kolik procent zaznamenaných nehod, které se staly na kolejích tramvaje, bylo smrtelných?

```python
on_tram_trails = clean[clean['situovani_nehody_na_komunikaci'] == 7]
deadly_on_tram_trails = clean[(clean['situovani_nehody_na_komunikaci'] == 7) & deadly]
tram_deadly = len(deadly_on_tram_trails) / len(on_tram_trails)
print(f"{tram_deadly*100}%")
```

### Vyberte časté výrobní značky vozidel (např. Škoda, BMW): ty které se účastnily alespoň padesáti nehod. Vykreslete graf, který pro každou častou výrobní značku ukáže procento nehod v zatáčce (vzhledem k celkovému počtu nehod vozidel této značky).
### Například: vozidla značky ŠKODA se účastnila 21421 nehod, z toho 2505 nehod bylo v zatáčce. 2505 z 21421 je 11,7%, graf tedy bude pro značku ŠKODA ukazovat 11,7%.

### Záznamy jako žádná z uvedených můžete pro jednoduchost považovat za samostatné značky.

```python
# In the next block will be same thing without for loop
d = {'brand': [], 'percentage of cornering accidents': []}
for i in range(100):
    all_accidents = len(clean[clean['vyrobni_znacka_motoroveho_vozidla'] == i])
    if (all_accidents > 50):
        in_kurve = len(clean[(clean['vyrobni_znacka_motoroveho_vozidla'] == i) & (clean['smerove_pomery'] == 3)])
        brand = decode_key['vyrobni_znacka_motoroveho_vozidla'][i]
        d['brand'].append(brand)
        d['percentage of cornering accidents'].append(in_kurve/all_accidents)
data = pd.DataFrame(data=d['percentage of cornering accidents'], index=d['brand'], columns=['percentage of cornering accidents'])
data.plot.bar()
```

```python
count = clean.groupby('vyrobni_znacka_motoroveho_vozidla').count()['identifikacni_cislo']
mask = count > 50

accidents_in_kurve = clean[clean['smerove_pomery'] == 3]
count_in_kurve = accidents_in_kurve.groupby('vyrobni_znacka_motoroveho_vozidla').count()['identifikacni_cislo']
percenntage = (count_in_kurve/count)[mask]
percenntage = percenntage.to_frame()

decode = pd.DataFrame(data=decode_key['vyrobni_znacka_motoroveho_vozidla'].values(), index=decode_key['vyrobni_znacka_motoroveho_vozidla'].keys(), columns=['brand'])

percenntage = percenntage.join(decode)
percenntage = percenntage.set_index('brand').rename(columns={"identifikacni_cislo": "percentage of cornering accidents"})
percenntage.plot.bar()
```
